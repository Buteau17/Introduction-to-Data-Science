#Boxplot

```


# Read the dataset
file_path = '2023 June Unemployment Rate by County (Percent).csv'
dataset = pd.read_csv(file_path)

# Extract the data you want to plot
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values

# Create a box plot for 'y' grouped by 'Region Name'
data = []
regions = set(x[1:, 1])  # Assuming the 'Region Name' column is the second column in 'x'

for region in regions:
    region_data = [float(y[i]) for i in range(len(y)) if x[i, 1] == region]
    data.append(region_data)

plt.boxplot(data, labels=regions)
plt.title("Box Plot of Unemployment Rate by Region")
plt.ylabel("Unemployment Rate")
plt.xticks(rotation=90)
plt.show()



```

#Scatter plot

```

# Filter out non-numeric values and convert the remaining values to floats
numeric_values = [float(value) for value in y if isinstance(value, (int, float)) or (isinstance(value, str) and value.replace('.', '', 1).isdigit())]

# Create a scatter plot of the unemployment rates
plt.scatter(range(len(numeric_values)), numeric_values, color='skyblue', edgecolor='black', alpha=0.6)

plt.title("Unemployment Rate Scatter Plot")
plt.xlabel("Data Point Index")
plt.ylabel("Unemployment Rate")

plt.show()
```

#QQ plot

```
import scipy.stats as stats
# Filter out non-numeric values and convert the remaining values to floats
numeric_values = [float(value) for value in y if isinstance(value, (int, float)) or (isinstance(value, str) and value.replace('.', '', 1).isdigit())]

# Calculate quantiles for the data
sorted_numeric_values = np.sort(numeric_values)
expected_quantiles = stats.norm.ppf(np.linspace(0, 1, len(numeric_values)))

# Create a QQ plot
plt.figure(figsize=(8, 6))
plt.scatter(expected_quantiles, sorted_numeric_values, color='skyblue', edgecolor='black', s=40)
plt.plot([np.min(expected_quantiles), np.max(expected_quantiles)], [np.min(expected_quantiles), np.max(expected_quantiles)], color='red', linestyle='--')
plt.title("QQ Plot")
plt.xlabel("Theoretical Quantiles")
plt.ylabel("Sample Quantiles")

plt.show()
```

#Density plot

```


# Filter out non-numeric values and convert the remaining values to floats
numeric_values = [float(value) for value in y if isinstance(value, (int, float)) or (isinstance(value, str) and value.replace('.', '', 1).isdigit())]

# Create a density plot for the data with outliers
plt.figure(figsize=(12, 6))
sns.kdeplot(numeric_values, color='skyblue', label='With Outliers')
plt.title("Unemployment Rate Density Plot (With Outliers)")
plt.xlabel("Unemployment Rate")
plt.ylabel("Density")

plt.show()


```

# Bar plot

```


# Calculate the total unemployment rates by region
data = {}
regions = set(x[1:, 1])  # Assuming the 'Region Name' column is the second column in 'x'

for region in regions:
    region_data = [float(y[i]) for i in range(len(y)) if x[i, 1] == region]
    total_unemployment_rate = sum(region_data)
    data[region] = total_unemployment_rate

# Convert the data dictionary into lists for plotting
region_names = list(data.keys())
total_rates = list(data.values())

# Create a bar plot for the total unemployment rates by region
plt.bar(region_names, total_rates)
plt.title("Total Unemployment Rate by Region")
plt.xlabel("Region Name")
plt.ylabel("Total Unemployment Rate")
plt.xticks(rotation=90)
plt.show()
```

# Histogram

```

numeric_values = [float(value) for value in y if isinstance(value, (int, float)) or (isinstance(value, str) and value.replace('.', '', 1).isdigit())]

# Create a histogram of the unemployment rates
plt.hist(numeric_values, bins=20, color='skyblue', edgecolor='black')

plt.title("Unemployment Rate Histogram")
plt.xlabel("Unemployment Rate")
plt.ylabel("Frequency")

plt.show()

```
